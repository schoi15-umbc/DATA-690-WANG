Since there are too many variables to explore, many of which are not interesting. Create a smaller data frame with the following 9 variables:


```
new_df = df[['UNITID', 'INSTNM', 'CITY', 'STABBR', 'ZIP', 'ADM_RATE', 'UGDS', 'TUITIONFEE_IN', 'MN_EARN_WNE_P10']]
new_df.head(5)
```

Definition of the 9 variables



*   UNITID = Unit ID for institution
*   INSTNM = Institution name
*   CITY = City
*   STABBR = State Postcode
*   ZIP = Zipcode
*   ADM_RATE = Admission Rate
*   UGDS = Enrollment of undergraduate certificate/degree-seeking students
*   TUITIONFEE_IN = In-state tuition and fees
*   MN_EARN_WNE_P10 = Mean earnings of students working and not enrolled 10 years after entry

```
#Find and display UMBC
UMBC= new_df[new_df['INSTNM'].str.contains("University of Maryland-Baltimore County")]
UMBC
```

```
#Find Univ. of Maryland College Park
UMCP = new_df[new_df['INSTNM'].str.contains("University of Maryland-College Park")]
UMCP
```

```
#Find and display Loyola College Maryland
LCM = new_df[new_df['INSTNM'].str.contains("Loyola University Maryland")]
LCM
```

```
#Find and display Johns Hopskins
JHU = new_df[new_df['INSTNM'].str.contains("Johns Hopkin")]
JHU
```

```
#Calculate the min, max, and average in-state tution of the above four institutions.
frames = [UMBC, UMCP, LCM, JHU]
result = pd.concat(frames)
```
